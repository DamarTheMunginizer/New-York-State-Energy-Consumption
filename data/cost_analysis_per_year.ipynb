```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year from the 'Date' column and create a new 'Year' column
df['Year'] = df['Date'].dt.year

# Drop non-numeric columns for summing, such as 'Date' and 'Year'
df_numeric = df.drop(columns=['Date', 'Year'])

# Group by the 'Year' column and sum the consumption for each county
annual_totals = df_numeric.groupby(df['Year']).sum()

# Optionally, you can create a new row with the total consumption for each year across all counties
annual_totals['Total_Consumption'] = annual_totals.sum(axis=1)

# Display the new DataFrame with the total for each year
pwr = annual_totals[['Total_Consumption']]
pwr
```

```python
# Define a function to pull the cost in for each year.
def calculate_and_save_energy_costs(pwr, output_filename="energy_costs.csv"):
    """
    Calculate costs for solar, wind, and hydro energy sources for annual consumption data
    and save the results to a CSV file.

    Parameters:
    - pwr (pd.DataFrame): Dataset with columns 'Year' and 'Total_Consumption' (in MW).
    - output_filename (str): Name of the output CSV file (default: 'energy_costs.csv').

    Returns:
    - pd.DataFrame: DataFrame containing the calculated costs for all years and energy sources.
    """
    # Define the energy sources and their specific parameters
    energy_sources = {
        "solar": {
            "install_cost_per_mw": 2000000,  # Average of $1,000,000 - $3,000,000 per MW
            "op_cost_per_mwh": 15000,       # Average of $10,000 - $20,000 per MW
            "lcoe": 40,                     #Average of $20 - $60 per MWh
            "lifespan": 25,                 # years
            "incentive_per_mw": .30      # 30% ITC on $1,000,000 average installation per MW
        },
        "wind": {
            "install_cost_per_mw": 1850000,  # Average of $1,200,000 - $2,500,000 per MW
            "op_cost_per_mwh": 20000,       # Average of $10,000 - $30,000 per MWh
            "lcoe": 45,                     # Average of $30 - $60 per MWh
            "lifespan": 22,                 # Average of 20-25 years
            "incentive_per_mw": .25      # PTC: $25 per MWh for 10 years, ~10,000 MWh/MW
        },
        "hydro": {
            "install_cost_per_mw": 3000000,  # Average of $1,000,000 - $5,000,000 per MW
            "op_cost_per_mwh": 22500,       #  Average of $15,000 - $30,000 per MWh
            "lcoe": 80,                     # Average of $40 - $120 per MWh
            "lifespan": 50,                 # Average of 40-60 years
            "incentive_per_mw": 0           # no specific federal incentive per MW
        }
    }

    results = pd.DataFrame()

    for energy_source, params in energy_sources.items():
        print(f"Processing {energy_source}...")

        # Calculate annual install cost for the energy source
        annual_install_cost_per_mw = (params["install_cost_per_mw"] - params["incentive_per_mw"]) / params["lifespan"]

        # Calculate costs for each year
        energy_cost = pwr["Total_Consumption"] * params["lcoe"]
        op_cost = pwr["Total_Consumption"] * params["op_cost_per_mwh"]
        annual_install_cost = pwr["Total_Consumption"] * annual_install_cost_per_mw
        total_cost = energy_cost + op_cost + annual_install_cost
        incentives = pwr["Total_Consumption"] * params["incentive_per_mw"]

        # Create a DataFrame for the current energy source
        source_results = pd.DataFrame({
            "Year": pwr["Year"],
            "Energy Source": [energy_source] * len(pwr),
            "Energy Cost ($)": energy_cost,
            "Operational Cost ($)": op_cost,
            "Annual Install Cost ($)": annual_install_cost,
            "Total Annual Cost ($)": total_cost,
            "Incentives ($)": incentives
        })

        # Append to the results
        results = pd.concat([results, source_results], ignore_index=True)

    # Save results to a CSV file
    results.to_csv(output_filename, index=False)
    print(f"Energy cost analysis saved to {output_filename}")

    return results
```

```python
# Run the function on the dataset
energy_costz = calculate_and_save_energy_costs(pwr, output_filename="energy_costz.csv")
```
